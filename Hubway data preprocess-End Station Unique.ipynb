```python
import pandas as pd
stations=pd.read_csv("C:/Users/prathik/Documents/Data Science Super Specialization/Tableau/Project/hubway_stations.csv")
```

```python
trips=pd.read_csv("C:/Users/prathik/Documents/Data Science Super Specialization/Tableau/Project/hubway_trips.csv")
```

```python
trips.head()
```

```python
result=trips
```

```python
unique_values=[]
for n in range(1,146):
    df=result.loc[result['end_statn'] == n]
    df=df.drop_duplicates(subset='strt_statn', keep='first', inplace=False)
    unique_values.append(df)
unique_values=pd.concat(unique_values)
```

```python
result = pd.merge(unique_values,
                 stations,
                 left_on='strt_statn',right_on='id')
result['origin_destination']='origin'
```

```python
result.head(2)
```

```python
result_1 = pd.merge(unique_values,
                 stations,
                 left_on='end_statn',right_on='id')
result_1['origin_destination']='destination'
result_1.head()
```

```python
final = pd.concat([result, result_1])
```

```python
final.head(2)
```

```python
final.sort_values(by='seq_id')
```

```python
list=final.columns.tolist()
drop_cols=['hubway_id', 'status_x', 'duration','start_date','end_date','bike_nr','subsc_type',
 'zip_code', 'birth_date', 'gender','municipal','status_y']
```

```python
final=final.drop(drop_cols,axis=1)
```

```python
final.head(2)
```

```python
final.sort_values(by='seq_id').head(2)
```

```python
bike_data=final.drop(['strt_statn','end_statn'],axis=1)
```

```python
bike_data.head(2)
```

```python
bike_data.to_csv("C:/Users/prathik/Documents/Data Science Super Specialization/Tableau/Project/bike_data_unique_end.csv")
```
